# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

## Answer

## Advantages and Disadvantages of Two Ways of Presenting Data

When plotting the number of visitors to a museum over two weeks, there are two common ways to show the data:

### 1. Separate Plots (One for Each Week)

#### Advantages:
- Easy to focus on each week individually.
- Clear visualization of the trend for each week.
- No overlapping lines, so it's less confusing.

#### Disadvantages:
- Hard to directly compare both weeks.
- Requires more space in the notebook or report.
- If axes are different, it may lead to misinterpretation.

---

### 2. Combined Plot (Both Weeks on the Same Graph)

#### Advantages:
- Easy to compare both weeks directly.
- Clear view of differences and similarities.
- Efficient use of space.

#### Disadvantages:
- Lines might overlap and cause confusion.
- Without labels or a legend, it can be hard to tell which line belongs to which week.
- Visualization may look messy if the data points are too close or cross over.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

years = np.arange(2007, 2015)

product_A = [400, 410, 420, 420, 400, 320, 300, 290]
product_B = [370, 390, 400, 410, 390, 310, 290, 280]
product_C = [np.nan, np.nan, np.nan, 150, 200, 210, 250, 210]  # Starts at 2010
product_D = [np.nan, np.nan, np.nan, np.nan, 180, 250, 220, 230]
product_E = [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, 90, 220]

plt.figure(figsize=(10, 6))
plt.plot(years, product_A, linewidth=3, color='grey')
plt.plot(years, product_B, linewidth=3, color='grey')
plt.plot(years, product_C, linewidth=3, color='grey')
plt.plot(years, product_D, linewidth=3, color='grey')
plt.plot(years[:len(product_E)], product_E, linewidth=3, color='grey')

plt.xlabel('Year')
plt.ylabel('Average Retail Product Price')

#  Add $ to Y-axis
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

plt.title('Average Retail Product Price per Year')

# Label lines (adjust year positions slightly to match visible line segments)
plt.text(2007.2, 400, 'A', fontsize=12, color='grey')
plt.text(2007.2, 370, 'B', fontsize=12, color='grey')
plt.text(2010.2, 100, 'C', fontsize=12, color='grey')
plt.text(2011.2, 180, 'D', fontsize=12, color='grey')
plt.text(2013.2, 90, 'E', fontsize=12, color='grey')

plt.ylim(0, 500)
plt.tight_layout()
plt.show()
```

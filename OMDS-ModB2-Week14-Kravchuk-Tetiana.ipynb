```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Set plot style
sns.set(style="whitegrid")

# Generate data
size = 12  # More data points
interval = 30
start = datetime.datetime(2024, 1, 1)
dates = [start + datetime.timedelta(days=interval) * i for i in range(size)]

np.random.seed(1)
frog = np.random.normal(size=size) + 0.3 * np.arange(size)
toad = np.random.normal(size=size) + 0.25 * np.arange(size)
salamander = np.random.normal(size=size) + 0.2 * np.arange(size)

# Create a DataFrame for clarity
df = pd.DataFrame({
    "Date": dates,
    "Frog": frog,
    "Toad": toad,
    "Salamander": salamander
})

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Frog"], label="Frog population (1000s)", marker='o')
plt.plot(df["Date"], df["Toad"], label="Toad population (1000s)", marker='s')
plt.plot(df["Date"], df["Salamander"], label="Salamander population (1000s)", marker='^')

plt.title("Amphibian Population Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Population (1000s)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```



####  Whats Good:
- The graph uses **time** on the x-axis, which is appropriate for showing trends over time.
- Each population (frog, toad, salamander) is labeled in the **legend**, which helps with identification.
- The data includes some **random variation**, making it feel more realistic.

---

####  What Could Be Improved:
- Only **5 data points** are shown, which makes the lines jagged and hard to interpret.
- The three lines **overlap**, making it difficult to tell them apart.
- There is **no title** or axis labels, so it's unclear what the graph is showing without reading the legend.
- The **dates** on the x-axis are not formatted clearly and may be hard to read.
- The **y-axis** doesnt show any units (even though the legend mentions "1000s").

---

####  Suggested Fixes:
- **Add more data points** to smooth out the trends and make patterns clearer.
- Add a **title** and labels for the **x-axis (Date)** and **y-axis (Population in 1000s)**.
- Use different **colors or markers** to better separate the lines visually.
- **Format the date labels** on the x-axis to improve readability.
- Use **gridlines** or a clearer background to make the chart easier to follow.

* Run t-tests and Mann Whitney U tests to compare different sets of numbers.
